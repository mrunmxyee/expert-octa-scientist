### Build a recommender system by using cosine simillarties score.

```python
import pandas as pd
```

```python
movie_data = pd.read_csv('Movie.csv')
movie_data
```

```python
movie_data['movie'].unique()
```

```python
movie_data['movie'].nunique()
```

```python
movie_data['userId'].unique()
```

```python
movie_data['userId'].nunique()
```

```python
pivot_data = pd.pivot_table( data = movie_data,values='rating',index='movie',columns='userId',).fillna(0)
pivot_data.columns = movie_data['userId'].unique()
pivot_data
```

```python
corr_metric = pivot_data.corr().round(3)
corr_metric
```

```python
euclideon_pivot_table = pivot_data.T
euclideon_pivot_table
```

```python
from sklearn.metrics import pairwise_distances
```

```python
euclideon_distance = pairwise_distances(X = euclideon_pivot_table,metric='euclidean')
euclideon_distance
```

```python
euclideon_distance = pd.DataFrame(data=euclideon_distance)
euclideon_distance.columns = movie_data['userId'].unique()
euclideon_distance.index   = movie_data['userId'].unique()
euclideon_distance.round(3)
```

```python
cosine_distance = 1 - pairwise_distances(X = euclideon_pivot_table,metric='cosine')
cosine_distance
```

```python
cosine_distance = pd.DataFrame(data=cosine_distance)
cosine_distance.columns = movie_data['userId'].unique()
cosine_distance.index   = movie_data['userId'].unique()
cosine_distance.round(3)
```
